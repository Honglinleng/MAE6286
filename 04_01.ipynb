```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['font.size'] = 16
```

```python
L = 1.0
nx = 51
dx = L/(nx-1)
alpha = 1.22e-3

x = np.linspace(0.0, L, num=nx)

T0 = np.zeros(nx)
T0[0] = 100.0
```

```python
def ftcs(T0, nt, dt, dx, alpha):
    T = T0.copy()
    sigma = alpha * dt / dx**2
    for n in range(nt):
        T[1:-1] = T[1:-1] + sigma * (T[:-2]-2 * T[1:-1] + T[2:])
  #     T[0] = 100.0
  #     T[-1] = 0.0
    return T
```

```python
nt = 100
sigma = 0.5
dt = sigma * dx**2 / alpha

T = ftcs(T0, nt, dt, dx, alpha)
```

```python
plt.plot(x, T, lw=3)
```

```python
nt = 100
sigma = 0.6
dt = sigma * dx**2 / alpha
T = ftcs(T0, nt, dt, dx, alpha)
```

```python
plt.plot(x, T, lw=3)
```

```python
def ftcs_mixed_bcs(T0, nt, dt, dx, alpha):
    T = T0.copy()
    sigma = alpha * dt / dx **2
    for n in range(nt):
        T[1:-1] = T[1:-1] + sigma * (T[:-2] - 2 * T[1:-1] + T[2:])
        #neumann condition with a zero-gradient value
        T[-1] = T[-2]
        
    return T
```

```python
nt = 1000
sigma = 0.5
dt = sigma * dx**2 / alpha
T = ftcs_mixed_bcs(T0, nt, dt, dx, alpha)
```

```python
plt.plot(x, T, lw=3)
```
